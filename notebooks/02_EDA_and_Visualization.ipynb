```python
# 02_eda_and_visuals.ipynb

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

df = pd.read_csv('../data/cleaned_telco.csv')

# Correlation heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# Distribution plot
numeric_cols = df.select_dtypes(include='number').columns[:4]
for col in numeric_cols:
    fig = px.histogram(df, x=col, title=f"Distribution of {col}")
    fig.show()

# Pie Charts
cat_cols = df.select_dtypes(include='object').columns[:2]
for col in cat_cols:
    fig = px.pie(df, names=col, title=f"{col} Distribution")
    fig.show()
```
